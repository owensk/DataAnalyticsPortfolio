### Narrative Rewrite

I see you there, young high school or university student, trying to find entertainment in a topic you've never heard about. Fear not. In this article, I'll be explaining the fascinating, productive, and valuable world of *Computational Fluid Dynamics*. You too can study the sciences and be a part of this important discipline. If you get really good at it, you may even be on one of the high profile F1 teams that do this as part of their daily workflow and design processes. And if you get really good at that, you might be able to be a part of a team that wins a world championship. That's right, you didn't mishear. You can be an F1 team member and call yourself a world champion, all for the small sacrifice of many years of math, science, and engineering training. But I've been blathering on too much already. Let me get to the modern day magic.

Behind every modern F1 team is a physics and math modeling team working to create the best aerodynamic design for their car. Top tier drivers deserve top tier engineering solutions and those require top tier engineering teams, and increasingly top tier computer science teams delivering high quality software solution to top tier physics modeling teams delivering incremental improvements to the car designs that deliver even faster speeds and higher performances. The work of these teams, all things being equal, is to shave of tenths of a second from the total lap time. Doing this iteratively, they'll stack those advantages and win more races, fingers crossed.

How do you go faster when everyone has a top of the line F1 race car? You generate more power, you deliver that power more efficiently to the ground, and you drive faster. With speed comes a problem of sticking to the road. F1 cars generate enormous amounts of downforce, which is just a fancy way of saying they push down on the ground with more force while than they do when they're sitting at rest. This happens because of *the wings* on these F1 cars. You read that right. **WINGS!** Modern day F1 cars have multiple wings, front and back, and some airflow modifying body designs, and these generate the enormous amounts of downforce required to run an F1 car at world championship speeds. 

This modeling is incredibly effective delivering consistent gains through the seaons and during the off season when working on the next season's car. The modeling is similar to the modeling down by aeronautical engineers when designing a new plane's airframe and wing design. Planes use lift and F1 cars use downforce but the methods of generating those forces are the same. Before math modeling and physics simulation, multiple forms of traditional aerodynamic testing were used. These methods are still used but are bolstered and their impact is improved because of design decision informed by modeling. 

When plane fly close to the ground, they cause *ground effects* which is increased lift due to airflow interaction with the surface of the earth. These ground effects are always in play with F1 cars because they only drive on the surface of the earth despite generating as much downforce as modern planes generate lift. Because of this more complicated modeling environment, the modeling of these systems does more than just modeling lift and drag. They model a component called **Wake Structure** which is related to how the air leave the surface of the F1 car during high velocity driving.

The problem is broken into two large regions, the front wing and the rear wing. The rear wing is frequently considered the most important because an overwhelming majority of the down force is generated here but the front wing's downforce is no less critical because it impacts steering control and effectiveness. Increasingly, the front wing is getting more attention as gains there are complicated and impact the performance of the rest of the F1 car's aerodynamic system. The front generates downforce, but it also is the primary source of wake structures that will impact the rest of the car's aerodynamics. 

The teams use wind tunnels to verify the computed expectations. Trust but verify. Both resources are restricted so doing highly effective work with both resources is integral to top tier performance. As technology, specifically computer performance, change the process of modeling has changed a lot. Machine learning is used to develop what are called *surrogate models* which are used to guide design space exploration. These surrogate models are trained on highly accurate historical wind tunnel data and pointed at the newest iteration of the racecar to effectively guess what should change where with a high level of machine learned skill. These are predictive algorithms that help prune the possible designs allowing the teams to focus on only the most likely of high impact design modifications.
